```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model, preprocessing
```

```python
df = pd.read_csv(r"C:\Users\zacha\神州\data\csv\car_evaluations.txt")
df.head()
```

```python
processor = preprocessing.LabelEncoder()
buying = processor.fit_transform(list(df['buying']))
maint = processor.fit_transform(list(df['maint']))
door = processor.fit_transform(list(df['door']))
persons = processor.fit_transform(list(df['persons']))
lug_boot = processor.fit_transform(list(df['lug_boot']))
safety = processor.fit_transform(list(df['safety']))
class1 = processor.fit_transform(list(df['class']))
print(buying)

predict = 'class1'
```

```python
x = list(zip(buying,maint,door,persons,lug_boot,safety))
y = list(class1)
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x,y, test_size = 0.2)
```

```python
print(x_train, y_test)
```

```python
model = KNeighborsClassifier(n_neighbors = 5)
model.fit(x_train, y_train)
accuracy = model.score(x_test, y_test)
print(accuracy)
```

```python
predicted = model.predict(x_test)
names = ['unacc', 'acc', 'good', 'vgood']
```

```python
for x in range(len(predicted)):
    print('Predicted: ', predicted[x], 'Data: ', x_test[x], 'Actual: ', y_test[x])
```

```python
for x in range(len(predicted)):
    print('Predicted: ', names[predicted[x]], 'Data: ', x_test[x], 'Actual: ', names[y_test[x]])
    n = model.kneighbors([x_test[x]], 9, True)
    print('N: ', n)
```
