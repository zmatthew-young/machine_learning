```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import BaggingClassifier

# Data
df = pd.read_csv(r"C:\Users\zacha\神州\data\csv\diabetes.csv")
x = df.drop('Outcome',axis=1)
y = df.Outcome

#Split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=9)

# Estimators
estimator_range = [2,4,6,8,10,12,14,16]

# Objects
models = []
scores = []

# For
for n_estimators in estimator_range:

    # Bagging
    model = BaggingClassifier(n_estimators=n_estimators,random_state=9)

    # Fit 
    model.fit(x_train, y_train)

    # Append 
    models.append(model)
    scores.append(accuracy_score(y_true=y_test,y_pred=model.predict(x_test)))

# Estimators by Score
plt.figure(figsize=(9,6))
plt.plot(estimator_range, scores)

# Plot
plt.title('Estimators by Score',fontsize=18)
plt.xlabel("n_estimators",fontsize=18)
plt.ylabel("score",fontsize=18)
plt.tick_params(labelsize=16)
plt.show()
```
