```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\zacha\神州\data\csv\income.csv")
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)']);
```

```python
km = KMeans(n_clusters=3)
km
```

```python
yp = km.fit_predict(df[['Age','Income($)']])
yp
```

```python
df['cluster'] = yp
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='blue')
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)']= scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])

df
```

```python
km = KMeans(n_clusters=3)
yp = km.fit_predict(df[['Age','Income($)']])
yp
```

```python
df['cluster'] = yp
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green',label='0')
plt.scatter(df2.Age,df2['Income($)'],color='red',label='1')
plt.scatter(df3.Age,df3['Income($)'],color='blue',label='2')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend(loc='upper left')
plt.show()
```
